Handling imbalanced data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, precision_recall_curve
```

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 5))
sns.distplot(x = df["Amount"], ax = ax[0], color = "b")
sns.distplot(x = df["Time"], ax = ax[1], color = "r")
                   
```

```python
fraud_df = df[df["Class"] == 1]
non_fraud_df = df[df["Class"] == 0]
resampled_df = resample(
    non_fraud_df,
    replace = False,
    n_samples = len(fraud_df),
    random_state = 10)
balanced_df = pd.concat([fraud_df, resampled_df])
balanced_df = balanced_df.sample(frac = 1, random_state = 10)
```

```python
fig, ax = plt.subplots(2, 1 ,figsize = (20, 16))
balanced_df_corr = balanced_df.corr()
df_corr = df.corr()
sns.heatmap(df_corr, cmap = "coolwarm", ax = ax[0])
sns.heatmap(balanced_df_corr, cmap = "coolwarm", ax = ax[1])
```

```python
fig, ax = plt.subplots(2, 4, figsize = (20, 10))
features  =["V2", "V4", "V11", "V19", "V10", "V12", "V14", "V16"]

for i, feature in  enumerate(features):
    row = i // 4
    col = i % 4
    sns.boxplot(x = "Class", y = feature, data = balanced_df, ax = ax[row, col], color = "red")
    ax[row, col].set_title(f"Boxplot of Class and {feature}")
    
plt.tight_layout()
```

```python
x_train = balanced_df.drop("Class", axis = 1)
y_train = balanced_df["Class"]

x_test = df.drop("Class", axis = 1)
y_test = df["Class"]

```

```python
pipeline = Pipeline([("scaler", RobustScaler()),
             ("model", LogisticRegression(class_weight = "balanced", random_state = 10, max_iter = 1000))])

pipeline.fit(x_train, y_train)

y_pred = pipeline.predict(x_test)
y_proba = pipeline.predict_proba(x_test)[:, -1]
y_proba
```

```python
print(classification_report(y_test, y_pred))

print(confusion_matrix(y_test, y_pred))

roc_score = roc_auc_score(y_test, y_proba)
print("ROC-AUC score:", roc_score) 
```

```python
from sklearn.metrics import precision_recall_curve, auc

# y_test = true labels, y_proba = predicted probabilities for positive class
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
auprc = auc(recall, precision)

print("AUPRC:", auprc)
```
